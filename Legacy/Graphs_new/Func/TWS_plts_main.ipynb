```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Palatino"],  # or Times, CMU, etc.
})

x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)

plt.plot(x, y, label=r"$\sin(x)$")
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.title(r"\textbf{LaTeX font test}")
plt.legend()
plt.show()
```

```python
import os, shutil

# Prefer the stable macOS TeX symlink (survives TeX Live version changes)
TEXBIN = "/Library/TeX/texbin"
if os.path.isdir(TEXBIN):
    os.environ["PATH"] = TEXBIN + ":" + os.environ.get("PATH", "")

# Sanity check (should not be None)
print("latex ->", shutil.which("latex"))
print("pdflatex ->", shutil.which("pdflatex"))
print("dvipng ->", shutil.which("dvipng"))
print("dvisvgm ->", shutil.which("dvisvgm"))

# Now it’s safe to use matplotlib + usetex
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Palatino"],  # or "TeX Gyre Pagella", "Times"
})

x = np.linspace(0, 2*np.pi, 200)
plt.plot(x, np.sin(x), label=r"$\sin(x)$")
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.title(r"\textbf{LaTeX font test}")
plt.legend()
plt.show()
```
