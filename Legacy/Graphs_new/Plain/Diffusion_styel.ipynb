```python
import numpy as np
import matplotlib.pyplot as plt

# === Global font size control ===
plt.rcParams.update({
    "font.size": 18,       # base font size
    "axes.titlesize": 20,  # title
    "axes.labelsize": 20,  # axis labels
    "xtick.labelsize": 16, # tick labels
    "ytick.labelsize": 16,
    "legend.fontsize": 16
})

# Domain for m
m = np.linspace(0, 1, 400)

# Functions
f1 = m
f2 = 1 - m
f3 = m * (1 - m)

# Figure (adjust figsize to taste, e.g. (5,4), (6,4), (7,5))
fig, ax = plt.subplots(figsize=(6, 4))

# Plot
ax.plot(m, f1, label=r"$m$", color="tab:blue", linewidth=2)
ax.plot(m, f2, label=r"$1-m$", color="tab:green", linewidth=2)
ax.plot(m, f3, label=r"$m(1-m)$", color="tab:red", linewidth=2)

# Highlight maximum of m(1-m)
m_max = 0.5
f3_max = f3[np.argmax(f3)]
ax.scatter(m_max, f3_max, color="black", zorder=5)
ax.text(m_max, f3_max + 0.03, r"max at $m=0.5$", ha="center", fontsize=16)

# Axes labels and legend
ax.set_xlabel(r"$m$ (ECM density)")
ax.set_ylabel("Value")
ax.set_title(r"Comparison of $m$, $1-m$, and $m(1-m)$")
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# === Global font size control ===
plt.rcParams.update({
    "font.size": 18,       # base font size
    "axes.titlesize": 20,  # title
    "axes.labelsize": 20,  # axis labels
    "xtick.labelsize": 16, # tick labels
    "ytick.labelsize": 16,
    "legend.fontsize": 16
})

# Domain for normalised m
mhat = np.linspace(0, 1, 400)

# Functions
f1 = mhat
f2 = 1 - mhat
f3 = mhat * (1 - mhat)

# Plot
fig, ax = plt.subplots(figsize=(6, 4))

ax.plot(mhat, f1, label=r"$\hat m$", color="tab:blue", linewidth=2)
ax.plot(mhat, f2, label=r"$1-\hat m$", color="tab:green", linewidth=2)
ax.plot(mhat, f3, label=r"$\hat m(1-\hat m)$", color="tab:red", linewidth=3)

# Vertical dashed line at max (mhat=0.5)
ax.axvline(0.5, color="black", linestyle="--", linewidth=1)

# Labels
ax.set_xlabel(r"$\hat m = m/M_x$")
ax.set_ylabel(r"$D(u,m)$")
ax.set_title(r"Normalised diffusion law: $\hat m(1-\hat m)$")

# Move legend outside
ax.legend(loc="center left", bbox_to_anchor=(1, 0.5))

ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_raw_m(Mx=1.0, figsize=(6,4), base_font=18):
    # --- Global font sizes ---
    plt.rcParams.update({
        "font.family": "Arial", 
        "font.size": base_font,
        "axes.titlesize": base_font + 2,
        "axes.labelsize": base_font + 2,
        "xtick.labelsize": base_font - 2,
        "ytick.labelsize": base_font - 2,
        "legend.fontsize": base_font - 2,
    })

    # Domain
    m = np.linspace(0, Mx, 400)

    # Functions
    f1 = m
    f2 = Mx - m
    f3 = m * (1 - m/Mx)

    # Plot
    fig, ax = plt.subplots(figsize=figsize)

    ax.plot(m, f1, label=r"$m$", color="tab:blue", linewidth=2)
    ax.plot(m, f2, label=r"$M_x - m$", color="tab:green", linewidth=2)
    ax.plot(m, f3, label=r"$m(1 - m/M_x)$", color="tab:red", linewidth=3)

    # Vertical dashed line at m = Mx/2
    ax.axvline(Mx/2, color="black", linestyle="--", linewidth=1)

    # Labels
    ax.set_xlabel(r"$m$ (ECM density)")
    ax.set_ylabel(r"$D(u,m)$")
    #ax.set_title(r"Functional form of the diffusion coefficient")
    # Main title (move higher with pad)
    ax.set_title("Functional form of diffusion", fontweight = 'bold',fontsize=18, pad=20)

    # Subtitle (smaller, just below main title)
    ax.text(0.5, .99, r"(plotted with $M_x=1$)", 
            transform=ax.transAxes,
            ha="center", va="bottom", fontsize=14)
    
    # Ticks for 0, Mx/2, Mx
    ax.set_xticks([0, Mx/2, Mx])
    ax.set_xticklabels([r"$0$", r"$M_x/2$", r"$M_x$"])

    # Move legend outside
    ax.legend(loc="center left", bbox_to_anchor=(1, 0.5))

    ax.grid(True, alpha=0.3)
    fig.tight_layout()
    plt.show()

# Example usage:
plot_raw_m(Mx=1.0)
```
