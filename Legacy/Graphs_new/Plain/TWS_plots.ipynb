```python
def plot_model_grid(
    models, m0_vals, lam_list,
    t_points=(60, 120, 180),
    yticks_mode="basic",
    show_arrows=True,
    arrow_len_frac=0.15,
    arrow_x_frac=0.70,
    y_red=0.80, y_blue=0.25,
    head_length=1.5, head_width=0.65, arrow_lw=2.3,
    show_speed_text=True,
    ceil_speed=False,
    row_ylabel=r"$u,\,m$",
    col_title_fmt=r"$m_0={m0}$",
    row_title_fmt=r"$\lambda={lam}$",
    fig_title=None,
    figsize=(13.6, 10.0),
    save=False, out_path="grid_models.pdf", dpi=600
):
    nrows, ncols = len(lam_list), len(m0_vals)
    fig, axes = plt.subplots(nrows, ncols, figsize=figsize, squeeze=False)
    fig.patch.set_facecolor("white")

    # Column headers
    for j, m0 in enumerate(m0_vals):
        axes[0, j].set_title(col_title_fmt.format(m0=m0), fontsize=18, pad=8)

    # iterate rows
    for i, lam in enumerate(lam_list):
        row_axes = axes[i, :]

        # --- Big row label (λ) on the left margin of the first panel ---
        left_ax = row_axes[0]
        lam_text = row_title_fmt.format(lam=_latex_sci(lam, pow10_threshold=100.0))
        left_ax.text(-0.25, 0.5, lam_text, transform=left_ax.transAxes,
                     ha="center", va="center", fontsize=22, fontweight="bold", rotation=90)

        # y-ticks setup
        shared_ylim = (0.0, 1.05) if yticks_mode != "splitplus" else (0.0, 1.25)
        yticks = {"basic":[0,0.5,1.0],
                  "split":np.arange(0,1.01,0.2),
                  "splitplus":np.arange(0,1.21,0.2)}.get(yticks_mode, [0,0.5,1.0])

        for j, m0 in enumerate(m0_vals):
            ax = axes[i, j]
            ax.grid(False)

            ax.set_ylim(shared_ylim)
            ax.set_yticks(yticks)
            if j > 0:
                ax.set_yticklabels([])
            else:
                ax.set_ylabel(row_ylabel, fontsize=18)

            if i < nrows - 1:
                ax.set_xticklabels([])
            ax.set_xlim(0, models[(m0, lam)].L)
            ax.tick_params(axis='x', labelsize=14)
            ax.tick_params(axis='y', labelsize=14)

            mdl = models[(m0, lam)]
            x, N_arr, M_arr, t_vec = mdl.x, mdl.N_arr, mdl.M_arr, mdl.times
            t_indices = _nearest_indices(t_vec, t_points)

            # estimate speed if needed
            c_val = None
            if show_speed_text:
                if getattr(mdl, "wave_speed", None) is None:
                    mdl.wave_speed, _, _ = mdl.estimate_wave_speed(
                        plot=False, target='N', threshold=0.5, band=(0.1, 0.9), spline_type='cubic'
                    )
                c_val = mdl.wave_speed

            # plot snapshots
            for t, tidx in zip(t_points, t_indices):
                ls = '--' if np.isclose(t, t_points[0]) else '-'
                ax.plot(x, N_arr[tidx], color="#6A00A8", linestyle=ls, lw=2.0)
                ax.plot(x, M_arr[tidx], color="#FF8C00", linestyle=ls, lw=2.0)

            # arrows
            if show_arrows:
                L = mdl.L
                ax.annotate('', xy=(arrow_x_frac*L + arrow_len_frac*L, y_red),
                            xytext=(arrow_x_frac*L, y_red),
                            arrowprops=dict(arrowstyle=f'->,head_length={head_length},head_width={head_width}',
                                            lw=arrow_lw, color="#FF8C00"))
                ax.annotate('', xy=(arrow_x_frac*L + arrow_len_frac*L, y_blue),
                            xytext=(arrow_x_frac*L, y_blue),
                            arrowprops=dict(arrowstyle=f'->,head_length={head_length},head_width={head_width}',
                                            lw=arrow_lw, color="#6A00A8"))

            # speed text
            if show_speed_text and (c_val is not None):
                if ceil_speed == "down":
                    c_str = f"{np.floor(c_val*100)/100:.2f}"
                elif ceil_speed == "up":
                    c_str = f"{np.ceil(c_val*100)/100:.2f}"
                else:
                    c_str = f"{c_val:.3g}"
                ax.text(0.03, 0.86, rf"$c={c_str}$", transform=ax.transAxes,
                        fontsize=18, ha="left", va="top")

    # bottom row x-labels
    for ax in axes[-1, :]:
        ax.set_xlabel(r"$x$", fontsize=18)

    plt.subplots_adjust(left=0.12, right=0.96, top=0.90, bottom=0.08,
                        wspace=0.08, hspace=0.24)
    if fig_title:
        fig.suptitle(fig_title, fontsize=20, y=0.98)

    if save:
        os.makedirs(os.path.dirname(out_path) or ".", exist_ok=True)
        root, ext = os.path.splitext(out_path)
        plt.savefig(root + ".pdf", bbox_inches="tight")
        plt.savefig(root + ".png", dpi=dpi, bbox_inches="tight")
    return fig, axes
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt

# ---- helpers ----
def _latex_sci(val, pow10_threshold=100.0):
    if val == 0:
        return "0"
    if abs(val) >= pow10_threshold:
        k = int(np.round(np.log10(val)))
        return r"10^{" + f"{k}" + r"}"
    return f"{val:g}"

def _nearest_indices(t_vec, t_points):
    return [int(np.argmin(np.abs(t_vec - t))) for t in t_points]

def _resolve_colors(color_scheme):
    """
    Returns (m_color, u_color) = (ECM, Tumour)
    color_scheme: 'opt1' | 'opt2' | 'opt3' | (m_color, u_color)
    """
    if isinstance(color_scheme, (list, tuple)) and len(color_scheme) == 2:
        return color_scheme[0], color_scheme[1]
    cs = str(color_scheme).lower()
    if cs == "opt1":
        return "#1f77b4", "#d62728"   # ECM blue, Tumour red
    if cs == "opt2":
        return "#1f77b4", "#ff8c00"   # ECM blue, Tumour orange
    if cs == "opt3":
        return "#6a00a8", "#ff8c00"   # ECM purple, Tumour orange
    # fallback
    return "#ff8c00", "#6a00a8"

# ---- main grid ----
def plot_model_grid(
    models, m0_vals, lam_list,
    t_points=(60, 120, 180),
    yticks_mode="basic",
    show_arrows=True,
    arrow_len_frac=0.15,
    arrow_x_frac=0.70,
    y_red=0.80, y_blue=0.25,        # vertical positions (we'll map to u/m colors below)
    head_length=1.5, head_width=0.65, arrow_lw=2.3,
    show_speed_text=True,
    ceil_speed=False,
    row_ylabel=r"$u,\,m$",
    col_title_fmt=r"$m_0={m0}$",
    row_title_fmt=r"$\lambda={lam}$",
    fig_title=r"Numerical solutions to the model for large $\lambda$",
    figsize=(13.6, 10.0),
    color_scheme="opt3",            # <- choose: 'opt1' | 'opt2' | 'opt3' | (m_color, u_color)
    save=False, out_path="Plots_Plain/grid_final", dpi=600
):
    """
    Build a rows x cols grid using solved model objects.
    Each model must provide: x, N_arr (tumour u), M_arr (ECM m), times, L, m0, k, estimate_wave_speed().
    Legend is placed at the bottom, no frame.
    """
    # colors
    m_color, u_color = _resolve_colors(color_scheme)

    nrows, ncols = len(lam_list), len(m0_vals)
    fig, axes = plt.subplots(nrows, ncols, figsize=figsize, squeeze=False)
    fig.patch.set_facecolor("white")

    # Column headers
    for j, m0 in enumerate(m0_vals):
        axes[0, j].set_title(col_title_fmt.format(m0=m0), fontsize=18, pad=8)

    # collect legend handles once (from the very first panel)
    legend_handles = None
    legend_labels  = [r"$u(x,t)$ (tumour)", r"$m(x,t)$ (ECM)"]

    for i, lam in enumerate(lam_list):
        # Big row label (λ) on the far left like a common y-label
        left_ax = axes[i, 0]
        lam_text = row_title_fmt.format(lam=_latex_sci(lam, pow10_threshold=100.0))
        left_ax.text(-0.25, 0.5, lam_text, transform=left_ax.transAxes,
                     ha="center", va="center", fontsize=22, fontweight="bold", rotation=90)

        # y ticks for this row
        shared_ylim = (0.0, 1.05) if yticks_mode != "splitplus" else (0.0, 1.25)
        yticks = {"basic":[0,0.5,1.0],
                  "split":np.arange(0,1.01,0.2),
                  "splitplus":np.arange(0,1.21,0.2)}.get(yticks_mode, [0,0.5,1.0])

        for j, m0 in enumerate(m0_vals):
            ax = axes[i, j]
            ax.grid(False)

            ax.set_ylim(shared_ylim)
            ax.set_yticks(yticks)
            if j > 0:
                ax.set_yticklabels([])
            else:
                ax.set_ylabel(row_ylabel, fontsize=18)

            if i < nrows - 1:
                ax.set_xticklabels([])
            mdl = models[(m0, lam)]
            ax.set_xlim(0, mdl.L)
            ax.tick_params(axis='x', labelsize=14)
            ax.tick_params(axis='y', labelsize=14)

            x, N_arr, M_arr, t_vec = mdl.x, mdl.N_arr, mdl.M_arr, mdl.times
            t_indices = _nearest_indices(t_vec, t_points)

            # estimate speed if asked
            c_val = None
            if show_speed_text:
                if getattr(mdl, "wave_speed", None) is None:
                    mdl.wave_speed, _, _ = mdl.estimate_wave_speed(
                        plot=False, target='N', threshold=0.5, band=(0.1, 0.9), spline_type='cubic'
                    )
                c_val = mdl.wave_speed

            # draw snapshots (first time dashed)
            h1 = h2 = None
            for t, tidx in zip(t_points, t_indices):
                ls = '--' if np.isclose(t, t_points[0]) else '-'
                # tumour u
                h1, = ax.plot(x, N_arr[tidx], color=u_color, linestyle=ls, lw=2.0)
                # ECM m
                h2, = ax.plot(x, M_arr[tidx], color=m_color, linestyle=ls, lw=2.0)

            # cache legend handles once
            if legend_handles is None and (h1 is not None) and (h2 is not None):
                legend_handles = [h1, h2]

            # arrows (match colors: u_color at y_red, m_color at y_blue)
            if show_arrows:
                L = mdl.L
                ax.annotate('', xy=(arrow_x_frac*L + arrow_len_frac*L, y_red),
                            xytext=(arrow_x_frac*L, y_red),
                            arrowprops=dict(arrowstyle=f'->,head_length={head_length},head_width={head_width}',
                                            lw=arrow_lw, color=u_color))
                ax.annotate('', xy=(arrow_x_frac*L + arrow_len_frac*L, y_blue),
                            xytext=(arrow_x_frac*L, y_blue),
                            arrowprops=dict(arrowstyle=f'->,head_length={head_length},head_width={head_width}',
                                            lw=arrow_lw, color=m_color))

            # speed text
            if show_speed_text and (c_val is not None):
                if ceil_speed == "down":
                    c_str = f"{np.floor(c_val*100)/100:.2f}"
                elif ceil_speed == "up":
                    c_str = f"{np.ceil(c_val*100)/100:.2f}"
                else:
                    c_str = f"{c_val:.3g}"
                ax.text(0.03, 0.86, rf"$c={c_str}$", transform=ax.transAxes,
                        fontsize=18, ha="left", va="top")

    # bottom x-labels
    for ax in axes[-1, :]:
        ax.set_xlabel(r"$x$", fontsize=18)

    # figure title (your requested text)
    if fig_title:
        fig = axes[0,0].figure
        fig.suptitle(fig_title, fontsize=20, y=1)

    # compact spacing, leave room at bottom for legend
    plt.subplots_adjust(left=0.12, right=0.96, top=0.92, bottom=0.12,
                        wspace=0.08, hspace=0.24)

    # bottom legend (no frame)
    if legend_handles is not None:
        fig.legend(legend_handles, legend_labels,
                   loc="lower center", ncol=2, frameon=False, fontsize=16,
                   bbox_to_anchor=(0.5, 0.02))

    if save:
        os.makedirs(os.path.dirname(out_path) or ".", exist_ok=True)
        plt.savefig(out_path + ".pdf", bbox_inches="tight")          # vector for LaTeX
        plt.savefig(out_path + ".png", dpi=dpi, bbox_inches="tight") # high-res raster
    return fig, axes
```

```python
# ---------- GAP & SUM GRIDS FOR m0 = 0.5 ----------
import os
import numpy as np
import matplotlib.pyplot as plt

# Build the dict from your existing variables (edit names if needed)
def _models_dict_m05():
    return {
        1e1:  model_05_10,
        1e2:  model_05_100,
        1e3:  model_05_1000,
        1e4:  model_05_10000,
        1e5:  model_05_100000,
        1e6:  model_05_1000000,
        1e7:  model_05_10000000,
        1e8:  model_05_100000000,
    }

def _nearest_indices(t_vec, t_points):
    return [int(np.argmin(np.abs(t_vec - t))) for t in t_points]

def _latex_sci_lam(val):
    # nice latex label: 10^k
    k = int(np.round(np.log10(val)))
    return rf"$\lambda = 10^{{{k}}}$"

def plot_sum_and_gap_grids_for_m05(
    lam_list=(1e1,1e2,1e3,1e4,1e5,1e6,1e7,1e8),
    t_points=(60, 120, 180),
    ncols=4,                          # 8 lambdas -> 2 rows x 4 cols by default
    colors_sum=("#1f77b4", "#1f77b4"),# same color for all times; style varies
    colors_gap=("#6a00a8", "#6a00a8"),
    lw=2.2,
    figsize_each=(16, 8),             # (width, height) per figure
    ylim_sum=(-0.05, 1.6),            # u+m ∈ [0,2], but 1.6 reads nicely here
    ylim_gap=(-0.05, 1.05),           # 1-(u+m) ∈ [0,1] ideally
    save=False, out_prefix="Plots_Plain/m05_sum_gap"
):
    """Create two figures: (A) u+m, (B) 1-(u+m), across the given lambdas."""
    models = _models_dict_m05()
    lam_list = list(lam_list)
    n = len(lam_list)
    nrows = int(np.ceil(n / ncols))

    # -------- Figure 1: u+m --------
    fig1, axes1 = plt.subplots(nrows, ncols, figsize=figsize_each, squeeze=False, sharex=False, sharey=True)
    axes1 = axes1.flatten()

    legend_handles = []
    legend_labels  = []
    for i, lam in enumerate(lam_list):
        ax = axes1[i]
        mdl = models.get(lam, None)
        if mdl is None:
            ax.set_axis_off()
            continue

        x, t, U, M = mdl.x, mdl.times, mdl.N_arr, mdl.M_arr
        tids = _nearest_indices(t, t_points)

        for k, (tt, tidx) in enumerate(zip(t_points, tids)):
            ls = "--" if k == 0 else "-"     # earliest dashed
            (ln,) = ax.plot(x, U[tidx] + M[tidx], ls, color=colors_sum[0], lw=lw)
            if i == 0:
                legend_handles.append(ln)
                legend_labels.append(rf"$t={int(t[tidx])}$")

        ax.set_xlim(0, mdl.L)
        ax.set_ylim(*ylim_sum)
        ax.set_title(_latex_sci_lam(lam), fontsize=18)
        if i % ncols == 0:
            ax.set_ylabel(r"$u(x,t)+m(x,t)$", fontsize=16)
        if i >= (nrows-1)*ncols:
            ax.set_xlabel(r"$x$", fontsize=16)
        ax.tick_params(axis='both', labelsize=12)
        ax.grid(False)

    # Empty panels (if any)
    for j in range(n, nrows*ncols):
        axes1[j].set_axis_off()

    # Figure title and legend
    fig1.suptitle(r"$u+m$ profiles across $\lambda$  (fixed $m_0=0.5$)", fontsize=22, y=0.98)
    if legend_handles:
        fig1.legend(legend_handles, legend_labels, loc="lower center",
                    ncol=len(t_points), frameon=False, fontsize=14, bbox_to_anchor=(0.5, 0.02))
    fig1.tight_layout(rect=[0.04, 0.06, 1, 0.95])

    if save:
        os.makedirs(os.path.dirname(out_prefix) or ".", exist_ok=True)
        fig1.savefig(out_prefix + "_sum.pdf", bbox_inches="tight")
        fig1.savefig(out_prefix + "_sum.png", dpi=300, bbox_inches="tight")

    # -------- Figure 2: 1 - (u+m) --------
    fig2, axes2 = plt.subplots(nrows, ncols, figsize=figsize_each, squeeze=False, sharex=False, sharey=True)
    axes2 = axes2.flatten()

    legend_handles = []
    legend_labels  = []
    for i, lam in enumerate(lam_list):
        ax = axes2[i]
        mdl = models.get(lam, None)
        if mdl is None:
            ax.set_axis_off()
            continue

        x, t, U, M = mdl.x, mdl.times, mdl.N_arr, mdl.M_arr
        tids = _nearest_indices(t, t_points)

        for k, (tt, tidx) in enumerate(zip(t_points, tids)):
            ls = "--" if k == 0 else "-"
            (ln,) = ax.plot(x, 1.0 - (U[tidx] + M[tidx]), ls, color=colors_gap[0], lw=lw)
            if i == 0:
                legend_handles.append(ln)
                legend_labels.append(rf"$t={int(t[tidx])}$")

        ax.set_xlim(0, mdl.L)
        ax.set_ylim(*ylim_gap)
        ax.set_title(_latex_sci_lam(lam), fontsize=18)
        if i % ncols == 0:
            ax.set_ylabel(r"$1-(u(x,t)+m(x,t))$", fontsize=16)
        if i >= (nrows-1)*ncols:
            ax.set_xlabel(r"$x$", fontsize=16)
        ax.tick_params(axis='both', labelsize=12)
        ax.grid(False)

    for j in range(n, nrows*ncols):
        axes2[j].set_axis_off()

    fig2.suptitle(r"Free-space profiles $\,1-(u+m)\,$ across $\lambda$  (fixed $m_0=0.5$)",
                  fontsize=22, y=0.98)
    if legend_handles:
        fig2.legend(legend_handles, legend_labels, loc="lower center",
                    ncol=len(t_points), frameon=False, fontsize=14, bbox_to_anchor=(0.5, 0.02))
    fig2.tight_layout(rect=[0.04, 0.06, 1, 0.95])

    if save:
        fig2.savefig(out_prefix + "_gap.pdf", bbox_inches="tight")
        fig2.savefig(out_prefix + "_gap.png", dpi=300, bbox_inches="tight")

    return (fig1, fig2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

def plot_um_and_gap_colormap(
    models_dict_fn,
    lam_list=(1e1,1e2,1e3,1e4,1e5,1e6,1e7,1e8),
    t_points=(0, 50, 100),
    cmap="plasma",              # <- your scheme
    lw=2.2,
    figsize=(13, 5.8),
    ylim_sum=(-0.05, 1.6),
    ylim_gap=(-0.05, 1.05),
    big_fonts=True,
    show_time_styles=True,      # dashed -> solid with increasing time
    save=False, out_prefix="Plots_Plain/m05_um_gap_colormap"
):
    """
    Single-panel plots for u+m and 1-(u+m) across many λ, coloured by log10(λ).
    """
    models = models_dict_fn()
    lam_list = list(lam_list)

    # Normalise λ for colormap
    log_lam = np.log10(np.asarray(lam_list, dtype=float))
    norm = plt.Normalize(vmin=log_lam.min(), vmax=log_lam.max())
    smap = cm.ScalarMappable(norm=norm, cmap=cmap)

    # Fonts
    if big_fonts:
        title_fs = 22; label_fs = 18; tick_fs = 14; cbar_fs = 16
    else:
        title_fs = 18; label_fs = 14; tick_fs = 12; cbar_fs = 12

    # Optional time→linestyle mapping
    if show_time_styles:
        # earliest dashed, mid dash-dot, latest solid (cycle if more)
        ls_cycle = ["--", "-.", "-"]
        linestyles = [ls_cycle[i % len(ls_cycle)] for i in range(len(t_points))]
    else:
        linestyles = ["-"] * len(t_points)

    # ---------- Figure 1: u+m ----------
    fig1, ax1 = plt.subplots(figsize=figsize)
    for lam in lam_list:
        mdl = models[lam]
        x, t, U, M = mdl.x, mdl.times, mdl.N_arr, mdl.M_arr
        tids = [int(np.argmin(np.abs(t - tp))) for tp in t_points]
        color = smap.to_rgba(np.log10(lam))
        for ls, tidx in zip(linestyles, tids):
            ax1.plot(x, U[tidx] + M[tidx], color=color, lw=lw, linestyle=ls)

    ax1.set_title(r"$u+m$ profiles  (fixed $m_0=0.5$)", fontsize=title_fs)
    ax1.set_xlabel(r"$x$", fontsize=label_fs)
    ax1.set_ylabel("Numerically estimated sum  $u+m$", fontsize=label_fs)
    ax1.set_ylim(*ylim_sum)
    ax1.tick_params(axis='both', labelsize=tick_fs)
    cbar1 = fig1.colorbar(smap, ax=ax1)
    cbar1.set_label(r"$\log_{10}(\lambda)$", fontsize=label_fs)
    cbar1.ax.tick_params(labelsize=cbar_fs)
    fig1.tight_layout()

    if save:
        fig1.savefig(out_prefix + "_sum_plasma.png", dpi=300, bbox_inches="tight")
        fig1.savefig(out_prefix + "_sum_plasma.pdf", bbox_inches="tight")

    # ---------- Figure 2: 1-(u+m) ----------
    fig2, ax2 = plt.subplots(figsize=figsize)
    for lam in lam_list:
        mdl = models[lam]
        x, t, U, M = mdl.x, mdl.times, mdl.N_arr, mdl.M_arr
        tids = [int(np.argmin(np.abs(t - tp))) for tp in t_points]
        color = smap.to_rgba(np.log10(lam))
        for ls, tidx in zip(linestyles, tids):
            ax2.plot(x, 1.0 - (U[tidx] + M[tidx]), color=color, lw=lw, linestyle=ls)

    ax2.set_title(r"$1-(u+m)$ profiles  (fixed $m_0=0.5$)", fontsize=title_fs)
    ax2.set_xlabel(r"$x$", fontsize=label_fs)
    ax2.set_ylabel(r"Free space $1-(u+m)$", fontsize=label_fs)
    ax2.set_ylim(*ylim_gap)
    ax2.tick_params(axis='both', labelsize=tick_fs)
    cbar2 = fig2.colorbar(smap, ax=ax2)
    cbar2.set_label(r"$\log_{10}(\lambda)$", fontsize=label_fs)
    cbar2.ax.tick_params(labelsize=cbar_fs)
    fig2.tight_layout()

    if save:
        fig2.savefig(out_prefix + "_gap_plasma.png", dpi=300, bbox_inches="tight")
        fig2.savefig(out_prefix + "_gap_plasma.pdf", bbox_inches="tight")

    return fig1, fig2
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# --------- helpers ---------
def _nearest_indices(t_vec, t_points):
    return [int(np.argmin(np.abs(t_vec - t))) for t in t_points]

def _models_dict_m01():
    # Adjust names if yours differ; keys must be the numeric lambdas you want to plot
    return {
        1e1:  model_01_10,
        1e2:  model_01_100,
        1e3:  model_01_1000,
        1e4:  model_01_10000,
        1e5:  model_01_100000,
        1e6:  model_01_1000000,
        1e7:  model_01_10000000,
        1e8:  model_01_100000000,
    }

def _models_dict_m05():
    return {
        1e1:  model_05_10,
        1e2:  model_05_100,
        1e3:  model_05_1000,
        1e4:  model_05_10000,
        1e5:  model_05_100000,
        1e6:  model_05_1000000,
        1e7:  model_05_10000000,
        1e8:  model_05_100000000,
    }

def _models_dict_m09():
    return {
        1e1:  model_09_10,
        1e2:  model_09_100,
        1e3:  model_09_1000,
        1e4:  model_09_10000,
        1e5:  model_09_100000,
        1e6:  model_09_1000000,
        1e7:  model_09_10000000,
        1e8:  model_09_100000000,
    }

# --------- main 1×3 gap figure ---------
def plot_gap_1x3_colormap(
    lam_list=(1e1,1e2,1e3,1e4,1e5,1e6,1e7,1e8),
    t_points=(0, 50, 100),
    cmap="plasma",
    lw=2.2,
    figsize=(18, 5.6),
    ylim_gap=(-0.05, 1.05),
    big_fonts=True,
    show_time_styles=True,   # dashed→solid with increasing time; set False for all-solid
    save=False, out_path="Plots_Plain/gap_1x3_m01_m05_m09"
):
    """
    1×3 panel of 1-(u+m) profiles for m0 = 0.1, 0.5, 0.9.
    Shared y-axis, single colorbar for log10(lambda).
    """
    # fonts
    if big_fonts:
        a = 3
        title_fs = 22; label_fs = 18; tick_fs = 14; cbar_fs = 16; panel_title_fs = 20
        title_fs = 22 + a; label_fs = 18 + a; tick_fs = 14 + a; cbar_fs = 16 + a; panel_title_fs = 20 + a
    else:
        title_fs = 18; label_fs = 14; tick_fs = 12; cbar_fs = 12; panel_title_fs = 16

    # linestyles for times
    if show_time_styles:
        ls_cycle = ["--", "-.", "-"]
        linestyles = [ls_cycle[i % len(ls_cycle)] for i in range(len(t_points))]
    else:
        linestyles = ["-"] * len(t_points)

    # prepare color mapping by log10(lambda)
    lam_arr = np.asarray(lam_list, dtype=float)
    log_lam = np.log10(lam_arr)
    norm = plt.Normalize(vmin=log_lam.min(), vmax=log_lam.max())
    smap = cm.ScalarMappable(norm=norm, cmap=cmap)

    # build models per m0
    panels = [
        (r"$m_0=0.1$", _models_dict_m01),
        (r"$m_0=0.5$", _models_dict_m05),
        (r"$m_0=0.9$", _models_dict_m09),
    ]

    fig, axes = plt.subplots(1, 3, figsize=figsize, sharey=True)
    for ax, (label_m0, models_fn) in zip(axes, panels):
        models = models_fn()
        # draw all lambdas for the requested times
        for lam in lam_list:
            mdl = models.get(lam, None)
            if mdl is None:
                continue  # skip missing runs silently
            x, t, U, M = mdl.x, mdl.times, mdl.N_arr, mdl.M_arr
            tids = _nearest_indices(t, t_points)
            color = smap.to_rgba(np.log10(lam))
            for ls, tidx in zip(linestyles, tids):
                ax.plot(x, 1.0 - (U[tidx] + M[tidx]), color=color, lw=lw, linestyle=ls)

        # panel cosmetics
        ax.set_title(label_m0, fontsize=panel_title_fs)
        ax.set_xlabel(r"$x$", fontsize=label_fs)
        ax.set_ylim(*ylim_gap)
        ax.tick_params(axis='both', labelsize=tick_fs)
        ax.grid(False)

    # only left panel gets ylabel (shared y)
    ## axes[0].set_ylabel(r"Free space $1-(u+m)$", fontsize=label_fs)
    axes[0].set_ylabel(r"Free space fraction $1-(u+m)$", fontsize=label_fs)

    # one shared colorbar
    cbar = fig.colorbar(smap, ax=axes.ravel().tolist(), shrink=0.9, pad=0.02)
    cbar.set_label(r"$\log_{10}(\lambda)$", fontsize=label_fs)
    cbar.ax.tick_params(labelsize=cbar_fs)

    fig.suptitle(r"Free space evolution during invasion for varying $\lambda$", 
             fontsize=title_fs, y=1.1)
    if save:
        import os
        os.makedirs(os.path.dirname(out_path) or ".", exist_ok=True)
        fig.savefig(out_path + ".png", dpi=300, bbox_inches="tight")
        fig.savefig(out_path + ".pdf", bbox_inches="tight")

    return fig, axes
```
